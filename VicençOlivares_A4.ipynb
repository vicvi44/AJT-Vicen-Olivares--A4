```python
def fibonacci(n):
    if n == 0:
        return [0]
    elif n == 1:
        return [0, 1]
    else:
        llista = [0, 1]
        for i in range(2, n):
            llista.append(llista[-1] + llista[-2])
        return llista

n = int(input("Introdueix el nombre de termes de Fibonacci: "))
print(fibonacci(n))
```

```python
from IPython.display import clear_output

def mostra_taulell(taulell, diccionari_jugadors, ):
    print("    " + "   ".join(str(i) for i in range(N)))
    for fila in range(N):
        contingut = " | ".join(diccionari_jugadors[taulell[fila][columna]] for columna in range(N))
        print(f" {fila}  {contingut}")
        if fila < N-1:
            print("   " + "---+" *(N-1) + "---")
    print()
    return


def demana_posicio():
    try:
        fila = int(input("Dona'm la fila on vols posar la peça (valor entre 0 " + str(N-1) + ")"))
        while fila < 0 or fila > N-1:
            fila = int(input("Siusplau, dona'm be la fila on vols posar la peça (valor entre 0 " + str(N-1) + ")"))
    except:
        return None, None
    try:
        columna = int(input("Dona'm la columna on vols posar la peça (valor entre 0 " + str(N-1) + ")"))
        while columna < 0 or columna > N-1:
            columna = int(input("Siusplau, dona'm be la columna on vols posar la peça (valor entre 0 i " + str(N-1) + ")"))
    except:
        return None, None
    if taulell[fila][columna] != 0:
        print("Aquesta posició ja està ocupada. Torna a provar")
        return demana_posicio()
    return fila, columna

def actualitza_taulell(taulell, fila, columna, jugador):
    taulell[fila][columna] = jugador
    return taulell

def canvi_jugador(jugador):
    return jugador*(-1)

def comprovar_victoria(taulell,jugador,):
    # Victòria per files
    for fila in range(0,N):
        suma = 0
        for columna in range(0,N):
            suma += taulell[fila][columna]
        if suma == N*jugador:
            return True
    # Victòria per columnes
    for columna in range(0,N):
        suma = 0
        for fila in range(0,N):
            suma += taulell[fila][columna]

        if suma == N*jugador:
            return True 
    # Victòria per diagonals
    suma = 0
    for diagonal in range(0,N):
        suma += taulell[diagonal][diagonal]            
    if suma == N*jugador:
        return True 
    # Victòria per diagonal
    suma = 0
    for diagonal in range(0,N):
        suma += taulell[diagonal][N-1-diagonal]            
    if suma == N*jugador:
        return True 
    
    return False
```
